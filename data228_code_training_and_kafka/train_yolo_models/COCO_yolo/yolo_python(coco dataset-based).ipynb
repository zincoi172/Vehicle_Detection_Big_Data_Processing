```python
import os
from datetime import datetime
from torch.utils.tensorboard import SummaryWriter
from ultralytics import YOLO

# 1. Automatically generate experiment name with timestamp
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
exp_name = f'yolov8_testlog_{timestamp}'
logdir = os.path.join('runs/train', exp_name)

# 2. Force initialize TensorBoard writer to ensure logs are created
os.makedirs(logdir, exist_ok=True)
writer = SummaryWriter(log_dir=logdir)
writer.add_scalar('init_scalar_force_log', 0, 0)  # Dummy scalar to trigger .tfevents file
writer.close()

# 3. Load YOLOv8 model (can be yolov8s.pt, yolov8m.pt, etc.)
model = YOLO("yolov8s.pt")

# 4. Train using built-in test dataset coco.yaml (auto-downloads coco128 if not found)
model.train(
    data="coco.yaml",      # Ultralytics built-in small dataset
    epochs=50,             # Number of training epochs
    imgsz=640,             # Image size
    batch=32,              # Batch size
    device=0,              # GPU index (0 = first GPU)
    project="runs/train",
    name=exp_name,
    verbose=True
)

print(f"\n Training complete! You can now view it in TensorBoard with:")
print(f"tensorboard --logdir runs/train/{exp_name}")
```

```python
from pycocotools.coco import COCO
import os

def convert_coco_to_yolo(json_path, output_dir):
    coco = COCO(json_path)
    os.makedirs(output_dir, exist_ok=True)

    for img_id in coco.getImgIds():
        img = coco.loadImgs(img_id)[0]
        ann_ids = coco.getAnnIds(imgIds=img['id'])
        anns = coco.loadAnns(ann_ids)

        label_path = os.path.join(output_dir, f"{img['file_name'].split('.')[0]}.txt")
        with open(label_path, "w") as f:
            for ann in anns:
                bbox = ann['bbox']
                x_center = (bbox[0] + bbox[2] / 2) / img['width']
                y_center = (bbox[1] + bbox[3] / 2) / img['height']
                width = bbox[2] / img['width']
                height = bbox[3] / img['height']
                category_id = ann['category_id'] - 1  # Adjust class index (COCO uses 1-based indexing)
                if category_id < 0 or category_id > 79:
                    continue  # Skip invalid class indices
                f.write(f"{category_id} {x_center} {y_center} {width} {height}\n")

# Paths to COCO annotation files
json_path_train = "C:/Users/jj111/OneDrive/Desktop/yolo/annotations/instances_train2017.json"
json_path_val = "C:/Users/jj111/OneDrive/Desktop/yolo/annotations/instances_val2017.json"

# Output label directories for YOLO format
output_train = "C:/Users/jj111/OneDrive/Desktop/yolo/labels/train2017"
output_val = "C:/Users/jj111/OneDrive/Desktop/yolo/labels/val2017"

# Convert and export YOLO-formatted labels
convert_coco_to_yolo(json_path_train, output_train)
convert_coco_to_yolo(json_path_val, output_val)

print("COCO annotations have been successfully converted to YOLO format!")
```
